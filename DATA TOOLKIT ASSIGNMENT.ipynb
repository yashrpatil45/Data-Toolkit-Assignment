# 1. What is NumPy, and why is it widely used in Python -->
---
NumPy (Numerical Python) is a powerful library for numerical computing in Python. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these data structures efficiently.

Why is NumPy Widely Used?
Performance

NumPy is significantly faster than Python lists because it is implemented in C and uses optimized algorithms for numerical computations.
It performs operations on entire arrays at once, leveraging vectorization, which eliminates slow Python loops.
Efficient Memory Usage

NumPy arrays consume less memory than Python lists since they store data in a contiguous block, reducing overhead.
Multi-Dimensional Arrays

The core data structure, ndarray, supports multi-dimensional arrays, which are essential for data science, machine learning, and scientific computing.
Comprehensive Mathematical Functions

NumPy includes built-in functions for linear algebra, statistics, Fourier transforms, and random number generation.
Interoperability

It integrates well with other libraries like Pandas, SciPy, Matplotlib, and TensorFlow, making it a fundamental package for data science and AI.
Broadcasting

NumPy allows operations on arrays of different shapes without explicit loops, making mathematical computations more efficient.
Indexing and Slicing

NumPy provides advanced indexing and slicing techniques that are more powerful and flexible than regular Python lists.

# 6.What are the differences between NumPy arrays and Python lists? -->
NumPy arrays and Python lists are both used for storing collections of data, but they have several key differences:

1. Performance
NumPy Arrays: Faster because they use a contiguous block of memory and optimized C and Fortran libraries for operations.
Python Lists: Slower because they are dynamically typed and store references to objects rather than actual values.
2. Memory Efficiency
NumPy Arrays: More memory-efficient because they store homogeneous data types in a compact way.
Python Lists: Less efficient because they store references to objects, which can take up more memory.
3. Type Homogeneity
NumPy Arrays: Must have elements of the same data type (e.g., all integers or all floats).
Python Lists: Can store mixed data types (e.g., integers, strings, and objects in the same list).
4. Operations & Broadcasting
NumPy Arrays: Support vectorized operations and broadcasting (e.g., element-wise addition, multiplication).

Python Lists: Require explicit loops or list comprehensions for element-wise operations.

# 13.Why is handling missing data important in Pandas -->
----

Ensures the dataset accurately represents the underlying data.
2. Prevents Errors in Computations

Proper handling (imputation or removal) ensures better predictive accuracy.
4. Enhances Data Visualization
Missing data can cause misleading trends in graphs.
Filling gaps ensures better interpretability in plots and charts.
5. Prepares Data for Further Processing
Many data pipelines require complete datasets.
Missing values can cause issues in merging, grouping, or transformations.

# 17. Explain the difference between apply() and map() in Pandas.
---
Difference Between apply() and map() in Pandas
Both apply() and map() are used for applying functions to Pandas Series or DataFrames, but they have key differences in their usage.

1. map() Method
Works only on Pandas Series (i.e., a single column).
Used for element-wise transformations.
Can take a function, dictionary, or Series for mapping.

2. apply() Method
Works on both Series and DataFrames.
Can apply a function row-wise (axis=1) or column-wise (axis=0).
More flexible than map().

# 21.Why is NumPy’s array slicing faster than Python’s list slicing?
---
NumPy’s array slicing is significantly faster and more efficient than Python’s list slicing due to several key reasons:

1. NumPy Uses Views Instead of Copies (Memory Efficiency)
When you slice a NumPy array, it does not create a new copy of the data. Instead, it creates a view that references the same memory.
In contrast, Python’s list slicing creates a new list, meaning data must be copied.

```python
#13.Create a 3D scatter plot using Plotly.
import plotly.graph_objects as go
import numpy as np

# Generate Sample Data
np.random.seed(42)
x = np.random.randint(1, 100, 20)  # 20 random X values
y = np.random.randint(1, 100, 20)  # 20 random Y values
z = np.random.randint(1, 100, 20)  # 20 random Z values

# Create 3D Scatter Plot
fig = go.Figure(data=[go.Scatter3d(
    x=x, y=y, z=z,
    mode='markers',
    marker=dict(size=6, color=z, colorscale='Viridis', opacity=0.8)
)])

# Customize Layout
fig.update_layout(
    title="3D Scatter Plot",
    scene=dict(
        xaxis_title="X Axis",
        yaxis_title="Y Axis",
        zaxis_title="Z Axis"
    )
)

# Show Plot
fig.show()
```
